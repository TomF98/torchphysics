# Hands-on: Create an image with a "T"

```python
# Create zero tensor (interprete as an image)
image = torch.zeros(size=(7,10))

# Vertical white stripe
image[:, 4:6] = 1

# Horizontal bar on top
image[:2, 2:-2] = 0.5 # equivalent to image[0:2, 2:-2] = 0.5

# Indexing with boolean tensor: Set values of all "T"-pixels to 0.7
is_T_pixel = image > 0
# Equivalent ways using logical operators: is_T_pixel = (image == 0.5) | (image == 1.0) logical "or" |
#                                          is_T_pixel = ~(image==0)                     logical negation ~
image[is_T_pixel] = 0.7

# Plot the image
plt.imshow(image, 'gray', vmin=0, vmax=1)
plt.colorbar(shrink=0.75)
plt.show()
```

# Exercise: Boolean indexing

```python
# Create image of shape (7,10) with random values between 0 and 1
random_2D_tensor = torch.rand(size=(7,10))

# TODO: Modify random_2D_tensor by boolean indexing:
#       Values >=0.7 should be set to 1, values<0.7 should be set to 0.
mask = random_2D_tensor >= 0.7
random_2D_tensor[mask] = 1.
random_2D_tensor[~mask] = 0.

# Plot the random image
plt.imshow(random_2D_tensor, 'gray', vmin=0, vmax=1)
plt.colorbar(shrink=0.75)
plt.show()
```
